```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
breast_cancer=load_breast_cancer()
print(breast_cancer)
```

```python
# print(breast_cancer.keys())
print(breast_cancer['feature_names'])
```

```python
cancer_frame=pd.DataFrame(np.c_[breast_cancer['data'],breast_cancer['target']],columns=np.append(breast_cancer['feature_names'],['target']))
print(cancer_frame)
```

```python
cancer_frame.to_csv('breast_cancer_dataframe.csv')
cancer_frame.head()
```

```python
x=cancer_frame.drop(['target'],axis=1)
# print(x)
y=cancer_frame['target']
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.40,random_state=0)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
print(x_train)
# print(x_test)
```

```python
#support vector classifier
from sklearn.svm import SVC
svm_classifier=SVC()
svm_classifier.fit(x_train,y_train)
```

```python
y_pred=svm_classifier.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_pred,y_test)
print(cm)
accuracy_score(y_pred,y_test)
classification_report(y_pred,y_test)
```
